# Problem 5

Find percentage of the prizes were awarded to 1, 2, 3, and 4 people? Meaning what percentage of 1/1, 1/2, 1/3, and 1/4 prize_shares are there?

```python
prize_share_counts = df['prize_share'].value_counts()
total_prizes = prize_share_counts.sum()

percentage_1 = (prize_share_counts.get('1/1', 0) / total_prizes) * 100
percentage_2 = (prize_share_counts.get('1/2', 0) / total_prizes) * 100
percentage_3 = (prize_share_counts.get('1/3', 0) / total_prizes) * 100
percentage_4 = (prize_share_counts.get('1/4', 0) / total_prizes) * 100

print(f"Percentage of prizes awarded to 1 person:", percentage_1, "%")
print(f"Percentage of prizes awarded to 2 people:", percentage_2, "%")
print(f"Percentage of prizes awarded to 3 people:", percentage_3, "%")
print(f"Percentage of prizes awarded to 4 people:", percentage_4, "%")
```

# Problem 6

Determine how many males and females won the prize. Does the sum of male and female prizes add up to the total entries? If not, why?

On the same graph make histograms for the numbber of males and females that won the prize each year. 

```python
gender_counts = df['sex'].value_counts()
total_entries = len(df)

print(f"Number of male laureates: {gender_counts.get('Male', 0)}")
print(f"Number of female laureates: {gender_counts.get('Female', 0)}")

if gender_counts.sum() == total_entries:
    print("Sum of male and female laureates adds up to the total entries.")
else:
    print("Sum of male and female laureates does not add up to the total entries.This is due to prizes being shared by 26 male and female colleagues, observed in problem 2.")

plt.figure(figsize=(10, 6))
df_male = df[df['sex'] == 'Male']
df_female = df[df['sex'] == 'Female']
plt.hist(df_male['year'], bins=range(1900, df['year'].max() + 1, 1), label='Male')
plt.hist(df_female['year'], bins=range(1900, df['year'].max() + 1, 1), label='Female')
plt.xlabel('Year')
plt.ylabel('Number of Nobel Prizes')
plt.title('Number of Nobel Prizes Awarded to Males and Females Each Year')
plt.legend()
plt.grid(axis='y')
plt.show()
```

# Problem 8

Determine the age that each of the prize winners was when they won the prize. Find the youngest and oldest to win the prize. Find the mean age to win the prize, and the standard deviation of it. Make a histogram of the age each person was when they won the prize.

### Hint: You can use Pandas DatetimeIndex method to extract the year from the date format yyyy-mm-dd
> pd.DatetimeIndex(date).year

Where *date* is given in a formate like year-month-day, the above will returen the year, hence the *.year* in the call above. Use 'shit+tab' on pandas.DatetimeIndex to learn more.

```python
df['birth_date'] = pd.to_datetime(df['birth_date'])

# Calculate age at the time of winning the prize as integers and handle NaN values by assigning 0 to prevent errors when ploting
df['age_at_win'] = (df['year'] - df['birth_date'].dt.year).fillna(0).astype(int)
filtered_df = df[df['age_at_win'] > 0]
youngest_winner = filtered_df.loc[filtered_df['age_at_win'].idxmin()]
oldest_winner = filtered_df.loc[filtered_df['age_at_win'].idxmax()]
mean_age = int(df['age_at_win'].mean())
std_dev_age = int(df['age_at_win'].std())

print("Youngest winner: ", youngest_winner['full_name'], "at age ", youngest_winner['age_at_win'])
print("Oldest winner:", oldest_winner['full_name'], "at age", oldest_winner['age_at_win'])
print("Mean age at winning:", mean_age)
print("Standard deviation of age at winning:", std_dev_age)

plt.figure(figsize=(10, 6))
plt.hist(df['age_at_win'], bins=range(1, filtered_df['age_at_win'].max() + 1, 1))
plt.xlabel('Age at Winning the Prize')
plt.ylabel('Number of Nobel Prizes')
plt.title('Age Distribution of Nobel Prize Winners')
plt.grid(axis='y')
plt.show()
```
